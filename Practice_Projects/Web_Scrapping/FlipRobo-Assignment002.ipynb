# QUESTION 1

```python
# Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
# have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
# jobs data.
# This task will be done in following steps:
# 1. First get the webpage https://www.naukri.com/
# 2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the
# location” field.
# 4. Then scrape the data for the first 10 jobs results you get.
# 5. Finally create a dataframe of the scraped data
# Note: All of the above steps have to be done in code. No step is to be done manually.

# Connect to the driver
driver=webdriver.Chrome(r"chromedriver.exe")

# Opening naukri page on automated chrome web browser
driver.get("https://www.naukri.com/")

# maximize the window's size
driver.maximize_window()

# entering designation and location as required in the question
designation=driver.find_element(By.CLASS_NAME, "suggestor-input")
designation.send_keys("Data Analyst")

location=driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")
location.send_keys("Bangalore")

# clicking on submit button
search=driver.find_element(By.CLASS_NAME, "qsbSubmit")
search.click()
time.sleep(3)

job_title=[]
job_location=[]
company_name=[]
experience_required=[]

# scrapping job title from the given page
title_tags=driver.find_elements(By.XPATH, '//a[@class="title fw500 ellipsis"]')
for i in title_tags[0:10]:
    job_title.append(i.text)
    
# scrapping job location from the given page
location_tags=driver.find_elements(By.XPATH, '//li[@class="fleft grey-text br2 placeHolderLi location"]')
for i in location_tags[0:10]:
    job_location.append(i.text)

# Scrapping company from the given page
name_tags=driver.find_elements(By.XPATH, '//a[@class="subTitle ellipsis fleft"]')
for i in name_tags[0:10]:
    company_name.append(i.text)

# Scrapping experience required from the given page
experience_tags=driver.find_elements(By.XPATH, '//span[@class="ellipsis fleft fs12 lh16 expwdth"]')
for i in experience_tags[0:10]:
    experience_required.append(i.text)
    
driver.close()   # closing the driver

# creating dataframe
df=pd.DataFrame({'Job title':job_title,'Job location':job_location,'Company name':company_name,'Exp required':experience_required})
df
```

```python
# Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
# 1. Brand
# 2. ProductDescription
# 3. Price
# The attributes which you have to scrape is ticked marked in the below image.
# To scrape the data you have to go through following steps:
# 1. Go to Flipkart webpage by url : https://www.flipkart.com/
# 2. Enter “sunglasses” in the search field where “search for products, brands and more” is written and
# click the search icon
# 3. After that you will reach to the page having a lot of sunglasses. From this page you can scrap the
# required data as usual.
# 4. After scraping data from the first page, go to the “Next” Button at the bottom other page , then
# click on it.
# 5. Now scrape data from this page as usual
# 6. Repeat this until you get data for 100 sunglasses.
# Note: That all of the above steps have to be done by coding only and not manually

# Connect to the driver
driver=webdriver.Chrome(r"chromedriver.exe")

# Opening flipkart page on automated chrome web browser
driver.get("https://www.flipkart.com/")
time.sleep(3)

# closing signin page
#close_signin=driver.find_element(By.XPATH, "/html/body/div[2]/div/div/button")
close_signin=driver.find_element(By.XPATH, '//button[@class="_2KpZ6l _2doB4z"]')
close_signin.click()

# entering “sunglasses” as required in the question
designation=driver.find_element(By.CLASS_NAME, "_3704LK")
designation.send_keys("sunglasses")

# click on search option
search=driver.find_element(By.XPATH, "/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button")
search.click()
time.sleep(3)

# scrapping details for 100 sun glasses
brand=[]
product_description=[]
price=[]
discount=[]
start=0
end=3
r=0
x=0
y=0
z=0
for page in range(start,end):
    brand_sunglasses=driver.find_elements(By.XPATH, '//div[@class="_2WkVRV"]')
    description=driver.find_elements(By.XPATH, '//a[@class="IRpwTa"]')
    price_details=driver.find_elements(By.XPATH, '//div[@class="_30jeq3"]')
    discount_details=driver.find_elements(By.XPATH, '//div[@class="_3Ay6Sb"]')
    for i in brand_sunglasses:
        brand.append(i.text)
        r=r+1
        if r==100:
            break
    for j in description:
        product_description.append(j.get_attribute('title'))
        x=x+1
        if x==100:
            break
    for k in price_details:
        price.append(k.text)
        y=y+1
        if y==100:
            break
    for l in discount_details:
        discount.append(l.text)
        z=z+1
        if z==100:
            break
    next_button=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(3)
    
print(len(brand),len(product_description),len(price),len(discount)) # printing the length of the list items

driver.close()  # closing the driver

#creating dataframe with the data as required in the question
df=pd.DataFrame({'Brand':brand,'Product Description':product_description,'Price':price,'Discount':discount})
df
```

```python
# Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
# The above task will be done in following steps:
# 1.First get the webpage https://www.azquotes.com/
# 2. Click on Top Quotes
# 3. Than scrap a) Quote b) Author c) Type Of Quotes

# Connect to the driver
driver=webdriver.Chrome(r"chromedriver.exe")

# Opening azquotes page on automated chrome web browser
driver.get("https://www.azquotes.com/")
time.sleep(2)

# maximize the window's size
driver.maximize_window()

# click on top quotes
search=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
search.click()
time.sleep(2)

# scrap a) Quote b) Author c) Type Of Quotes
quote=[]
author=[]
type_of_quotes=[]

start=0
end=10

for page in range(start,end):
    quote_data=driver.find_elements(By.XPATH, '//a[@class="title"]')
    author_data=driver.find_elements(By.XPATH, '//div[@class="author"]')    
    type_of_quotes_details=driver.find_elements(By.XPATH, '//div[@class="tags"]') 

    for i in quote_data:    
        quote.append(i.text)  
    for k in type_of_quotes_details:    
        type_of_quotes.append(k.text)       
    for l in author_data:    
        author.append(l.text)
    if page <= 8:
        next_button=driver.find_element(By.XPATH,'(//li[@class="next"]//a)[1]')
        next_button.click()
        time.sleep(1)

print(len(quote),len(author),len(type_of_quotes))  # printing the length of the list items

driver.close()  # closing the driver

#creating dataframe with the data as required in the question
df=pd.DataFrame({'Quote':quote,'Author':author,'Type of Quotes':type_of_quotes})
df
```

```python
# Q9: Write a python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead,
# Term of office, Remarks) from https://www.jagranjosh.com/.
# This task will be done in following steps:
# 1. First get the webpage https://www.jagranjosh.com/
# 2. Then You have to click on the GK option
# 3. Then click on the List of all Prime Ministers of India
# 4. Then scrap the mentioned data and make the DataFrame.

# Connect to the driver
driver=webdriver.Chrome(r"chromedriver.exe")

# Opening https://www.jagranjosh.com/ page on automated chrome web browser
driver.get("https://www.jagranjosh.com/")
time.sleep(3)

# maximize the window's size
driver.maximize_window()

# click on GK option
search=driver.find_element(By.XPATH,'/html/body/div/div[1]/div/div[1]/div/div[6]/div/div[1]/header/div[3]/ul/li[9]/a')
search.click()
time.sleep(3)

# click on the List of all Prime Ministers of India
search_PM=driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div[2]/div/div[10]/div/div/ul/li[2]/a')
search_PM.click()
time.sleep(3)

# scrap Name, Born-Dead, Term of office, Remarks
Name=[]
Born_Dead=[]
Term_of_office_i=[]
Term_of_office=[]
Remark=[]

All_data=driver.find_elements(By.TAG_NAME, 'td')

k=1
l=2
m=3
n=4
x=0
kk=0

for i in All_data:
    if(x==k):
        Name.append(i.text)
        k=k+5
    if(x==l):
        Born_Dead.append(i.text)
        l=l+5
    if(x==m):
        Term_of_office_i.append(i.text)
        res_1=Term_of_office_i[kk].split()
        res_2=" ".join(res_1)
        Term_of_office.append(res_2)
        kk=kk+1
        m=m+5
    if(x==n):
        Remark.append(i.text)
        n=n+5
    x=x+1
    if(x==90):
        break

    
print(len(Name),len(Born_Dead),len(Term_of_office),len(Remark))   # printing the length of the items

driver.close()  # closing the web driver

#creating dataframe with the data as required in the question
df=pd.DataFrame({'Name':Name,'Born Dead':Born_Dead,'Term of office':Term_of_office,'Remarks':Remark})
df
```

```python
# Q10: Write a python program to display list of 50 Most expensive cars in the world (i.e.
# Car name ,Description and Price) from https://www.motor1.com/
# This task will be done in following steps:
# 1. First get the webpage https://www.motor1.com/
# 2. Then You have to click on the List option from Dropdown menu on leftside.
# 3. Then click on 50 most expensive carsin the world..
# 4. Then scrap the mentioned data and make the

# Connect to the driver
driver=webdriver.Chrome(r"chromedriver.exe")

# Opening https://www.motor1.com/ page on automated chrome web browser
driver.get("https://www.motor1.com/")
time.sleep(3)

# maximize the window's size
driver.maximize_window()

# click on LEFT MOST option
left=driver.find_element(By.XPATH,'/html/body/div[3]/div[2]/div/div/div[1]/div')
left.click()
time.sleep(3)

# click on FEATURES option
features=driver.find_element(By.XPATH,'/html/body/div[4]/div[1]/div[3]/ul/li[5]/button')
features.click()
time.sleep(3)

# click on LISTS option
lists=driver.find_element(By.XPATH,'/html/body/div[4]/div[1]/div[3]/ul/li[6]/ul/li[1]/a')
lists.click()
time.sleep(3)

# click on MOST EXPENSIVE CARS IN THE WORLD
expensive_cars=driver.find_element(By.XPATH,'/html/body/div[3]/div[9]/div[1]/div[1]/div/div/div[2]/div/div[1]/h3/a')
expensive_cars.click()
time.sleep(3)

car_name=[]
car_description=[]
car_price=[]

# scrapping Car name ,Description and Price

car_name_data=driver.find_elements(By.XPATH, '//h3[@class="subheader"]')  # scrapping car name data
All_data=driver.find_elements(By.TAG_NAME, 'p')  # scrapping both description and price data

k=4
l=5
x=0
y=0

for j in car_name_data:       # scrapping car name data
    car_name.append(j.text)
    y=y+1
    if(y==50):
        break

for i in All_data:            # scrapping both description and price data
    if(x==k):
        car_price.append(i.text)     # scrapping car price details
        k=k+2
    if(x==l):
        car_description.append(i.text)   # scrapping car description details
        l=l+2
    x=x+1
    if(x==104):
        break
    
print(len(car_name),len(car_description),len(car_price))   # printing the lenght of the items

driver.close()   # close the web driver

#creating dataframe with the data as required in the question
df=pd.DataFrame({'Car Name':car_name,'Description':car_description,'Price':car_price})
df
```

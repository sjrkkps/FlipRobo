# QUESTION 1

```python
# 1. Scrape the details of most viewed videos on YouTube from Wikipedia.
# Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:
# A) Rank
# B) Name
# C) Artist
# D) Upload date
# E) Views

web_page = requests.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos") #creating request object
bsoup=BeautifulSoup(web_page.content)  # using beautiful soup to extract the data. #creating soup object
rows = bsoup.find_all('tr')  # Extract all the rows with 'tr' tag
contents = []
for i in rows:
    columns=i.find_all('td')  # Extract all the cells with 'td' tag from each row and store in a list
    columns=[x.text.strip() for x in columns] # Scrapping each element from the list
    contents.append(columns) # Appending list "contents" with each scrapped element
# creating the dataframe with contents
df1 = pd.DataFrame(data=contents) # dataframe with "contents"
df2 = df1.iloc[2:32 ,0:5]  # extracting the required data per question
# Defining required column names
df2.columns =['Rank','Video Name', 'Uploader', 'Views (billions)','Upload Date']
# Displaying the dataframe
df2
```

# QUESTION 2

```python
# 2. Scrape the details team India’s international fixtures from bcci.tv. Url = https://www.bcci.tv/.
# You need to find following details:
# A) Match title (I.e. 1st ODI)
# B) Series
# C) Place
# D) Date
# E) Time
# Note: - From bcci.tv home page you have reach to the international fixture page through code.

# Connect to the driver
driver=webdriver.Chrome(r"chromedriver.exe")

# Opening web page on automated chrome web browser
driver.get("https://www.bcci.tv/")
time.sleep(3)  # 3 seconds sleep

# maximize the window's size
driver.maximize_window()

#delete the cookies  
driver.delete_all_cookies()

# click on "INTERNATIONAL" option
search=driver.find_element(By.XPATH, "/html/body/nav/div[1]/div[2]/ul[1]/li[2]/a")
search.click()
driver.implicitly_wait(5) # seconds

# click on "ALL TEAMS" option
search=driver.find_element(By.XPATH, '//div[@class="col-lg-2 col-md-2 col-sm-12"]//div[@class="customSelecBox"]')
search.click()
driver.implicitly_wait(5) # seconds

# click on "INDIA" option
# search=driver.find_element(By.XPATH, '//div[@class="col-lg-2 col-md-2 col-sm-12"]//div[@class="cSBListItems ng-binding ng-scope"][2]')
search=driver.find_element(By.XPATH, '//div[@class="col-lg-2 col-md-2 col-sm-12"]//div[contains(text(),"india")]')
search.click()
driver.implicitly_wait(5) # seconds

for _ in range(3): # scroll the page for 3 times
    driver.execute_script("window.scrollBy(0,500)") # scroll the page by 500 pixes each time
    time.sleep(2) # wait for 2 seconds to go to next command 

# defining empty list items as required
Match_title=[]
Series=[]
Place=[]
Date=[]
Time=[]

start=0
end=1
for page in range(start,end): # for loop for scrapping 3 pages
    # scrapping Match Title Names
    title=driver.find_elements(By.XPATH, '//span[@class="matchOrderText ng-binding ng-scope"]') # finding all elements of MATCH TITLES
    try:
        for i in title: # for loop with no. of match titles
            Match_title.append(i.text.strip(" ")[:-1]) # appending the match titles to the list
    except NoSuchElementException: # Error handling exception
        Match_title.append("_")  # Use "_" for missing titles
        
    # scrapping Series details
    series_info=driver.find_elements(By.XPATH, '//span[@class="ng-binding"]') 
    try:
        for i in series_info: 
            Series.append(i.text.strip(" ")) 
    except NoSuchElementException: 
        Series.append("_")  
        
    # scrapping Place details
    Place_info=driver.find_elements(By.XPATH, '//div[@class="fix-place ng-binding ng-scope"]') 
    result1=[]
    try:
        for i in Place_info: 
            result1.append(i.text.strip(" "))
            result2=[i.split("-") for i in result1]
            result3 = [item for l in result2 for item in l]
            Place=[i for i in result3 if result3.index(i)%2!=0]
    except NoSuchElementException: 
        Place.append("_") 
        
    # scrapping Date details
    Date_info=driver.find_elements(By.XPATH, '//h5[@class="ng-binding"]') 
    try:
        for i in Date_info: 
            Date.append(i.text.strip(" ")) 
    except NoSuchElementException: 
        Date.append("_")  
        
    # scrapping Time details
    Time_info=driver.find_elements(By.XPATH, '//h5[@class="text-right ng-binding"]') 
    try:
        for i in Time_info: 
            Time.append(i.text.strip(" ")) 
    except NoSuchElementException: 
        Time.append("_") 
        
driver.close()  # closing the driver

# printing lengths of all items    
print(len(Match_title),len(Series),len(Place),len(Date),len(Time))

# creating a data frame with all the information scrapped
df=pd.DataFrame({'Match title':Match_title,'Series':Series,'Place':Place,'Date':Date,'Time':Time})

# displaying dataframe
df
```

```python
# 9. Scrape the details most watched tv series of all time from imdb.com. Url = https://www.imdb.com/list/ls095964455/
# You have to find the following details:
# A) Name
# B) Year span
# C) Genre
# D) Run time
# E) Ratings
# F) Votes

# Connect to the driver
driver=webdriver.Chrome(r"chromedriver.exe")

# Opening web page on automated chrome web browser
driver.get("https://www.imdb.com/list/ls095964455/")
time.sleep(3)  # 3 seconds sleep

# maximize the window's size
driver.maximize_window()

#delete the cookies  
driver.delete_all_cookies()

# scrapping details as required in the question
Name=[]
Year_Span=[]
Genre=[]
Run_Time=[]
Ratings=[]
Votes=[]

# scrapping Name
name_info=driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]/a') 
try:
    for i in name_info:
        Name.append(i.text.strip(" ")) 
except NoSuchElementException:
    Name.append("_")

# scrapping Year_Span
Year_Span_info=driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]//span[@class="lister-item-year text-muted unbold"]') 
try:
    for i in Year_Span_info:
        Year_Span.append(i.text.strip(" ")) 
except NoSuchElementException:
    Year_Span.append("_")
    
# scrapping Genre
Genre_info=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/p[1]/span[5]') 
try:
    for i in Genre_info:
        Genre.append(i.text.strip(" ")) 
except NoSuchElementException:
    Genre.append("_")
    
# scrapping Run_Time
Run_Time_info=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/p[1]/span[3]') 
try:
    for i in Run_Time_info:
        Run_Time.append(i.text.strip(" ")) 
except NoSuchElementException:
    Run_Time.append("_")
    
# scrapping Ratings
Ratings_info=driver.find_elements(By.XPATH,'//div[@class="ipl-rating-star small"]/span[2]') 
try:
    for i in Ratings_info:
        Ratings.append(i.text.strip(" ")) 
except NoSuchElementException:
    Ratings.append("_")
    
# scrapping Votes
Votes_info=driver.find_elements(By.XPATH,'//p[@class="text-muted text-small"]//span[@name="nv"]') 
try:
    for i in Votes_info:
        Votes.append(i.text.strip(" ")) 
except NoSuchElementException:
    Votes.append("_")
    
driver.close()  # closing the driver
    
# printing lengths of all items    
print(len(Name),len(Year_Span),len(Genre),len(Run_Time),len(Ratings),len(Votes))

# creating a data frame with all the information scrapped
df=pd.DataFrame({'Name':Name,'Year Span':Year_Span,'Genre':Genre,'Run Time':Run_Time,'Ratings':Ratings,'Votes':Votes})

#Displaying the dataframe
df
```

```python
# 10. Details of Datasets from UCI machine learning repositories. Url = https://archive.ics.uci.edu/
# You have to find the following details:
# A) Dataset name
# B) Data type
# C) Task
# D) Attribute type
# E) No of instances
# F) No of attribute
# G) Year
# Note: - from the home page you have to go to the ShowAllDataset page through code.

# Connect to the driver
driver=webdriver.Chrome(r"chromedriver.exe")

# Opening web page on automated chrome web browser
driver.get("https://archive.ics.uci.edu/ml/index.php")
time.sleep(3)  # 3 seconds sleep

# maximize the window's size
driver.maximize_window()

#delete the cookies  
driver.delete_all_cookies()

# click on "View All Sets" option
economy=driver.find_element(By.XPATH, '/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b').click()
driver.implicitly_wait(5) # seconds

# defining empty list items as required
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attributes=[]
Year=[]
result1=[]

# scrapping the required details
All_info=driver.find_elements(By.XPATH, '/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td') # finding all elements of table
try:
    for i in All_info: # for loop with all information
        result1.append(i.text.strip(" ")) # scrapping all details of table
        le=len(result1) # length of list items
        Dataset_name=[result1[i] for i in range(0,le,7)] # scrapping "Dataset_name" details
        Data_type=[result1[i+1] for i in range(0,le-1,7)] # scrapping "Data_type" details
        Task=[result1[i+2] for i in range(0,le-2,7)] # scrapping "Default Task" details
        Attribute_type=[result1[i+3] for i in range(0,le-3,7)] # scrapping "Attribute_type" details
        No_of_instances=[result1[i+4] for i in range(0,le-4,7)] # scrapping "Instances" details
        No_of_attributes=[result1[i+5] for i in range(0,le-5,7)] # scrapping "Attributes" details
        Year=[result1[i+6] for i in range(0,le-6,7)] # scrapping "Year" details
except NoSuchElementException: # Error handling exception
    print("Error")
    
driver.close()  # closing the driver

# printing lengths of all items    
print(len(Dataset_name),len(Data_type),len(Task),len(Attribute_type),len(No_of_instances),len(No_of_attributes),len(Year))
le=len(Dataset_name) # length of the list

# creating a data frame with all the information scrapped
df=pd.DataFrame({'Dataset Name':Dataset_name[1:le],'Data Types':Data_type[1:le],'Task':Task[1:le],'Attribute Types':Attribute_type[1:le],'No of Instances':No_of_instances[1:le],'No of Attributes':No_of_attributes[1:le],'Year':Year[1:le]})

#Displaying the dataframe
df
```
